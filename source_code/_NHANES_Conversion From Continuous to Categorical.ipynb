```python
def agecontinuous (age):
    if ((age >= 20.0) and (age <= 30.0)):
        val = 1
    elif ((age > 30.0) and (age <= 40.0)):
        val = 2
    elif ((age > 40.0) and (age <= 50.0)):
        val = 3
    elif ((age > 50.0) and (age <= 60.0)):
        val = 4
    elif ((age > 60.0) and (age <= 70.0)):
        val = 5
    elif ((age > 70.0) and (age <= 80.0)):
        val = 6
    return val
```

```python
#Underweight = <18.5 0
#Normal weight = 18.5–24.9  1
#Overweight = 25–29.9 2
#Obesity = BMI of 30 or greater 3
def bmicontinuous (bmi):
    if (bmi < 18.5):
        ret = 1
    elif ((bmi >= 18.5) and (bmi <= 24.9)):
        ret = 2
    elif ((bmi >= 25) and (bmi <= 29.9)):
        ret = 3
    #elif (bmi >= 30):
    else:
        ret = 4
    return ret
```

```python
df['AGE'] = df.apply(lambda x : agecontinuous(x['RIDAGEYR']),axis=1)
```

```python
df['BMI'] = df.apply(lambda x : bmicontinuous(x['BMXBMI']),axis=1)
```

```python
df_new = pd.DataFrame()
df_new['SEQN']=df['SEQN']
```

```python
df_new['GENDER']=df['RIAGENDR'].astype(int)
df_new['AGERANGE'] = df.apply(lambda x : agecontinuous(x['RIDAGEYR']),axis=1)
df_new['RACE'] = df['RIDRETH1'].astype(int)
df_new['BMIRANGE'] = df.apply(lambda x : bmicontinuous(x['BMXBMI']),axis=1)
df_new['KIDNEY'] = df['KIQ022'].astype(int)
df_new['SMOKE'] = df['SMQ020'].astype(int)
df_new['DIABETES'] = df['DIQ010'].astype(int)
df_new['HYPCLASS']=df['HYPCLASS']
```
