```python
#Claculation of KIQ022 for dataset 2015-2016

#MDRD STUDY EQUATION
#GFR (mL/min/1.73 m2) = 175 × (Scr)^-1.154 × (Age)^-0.203 × (0.742 if female) × (1.212 if African American)
#SCr (standardized serum creatinine) = mg/dL, age = years

#Prevalence of CKD was defined by 
#https://renal.org/information-resources/the-uk-eckd-guide/ckd-stages/
#Glomerular Filtration Rate (eGFR) < 60 ml/min/1.73 m^2  or albumin - creatinine ratio >= 30 mg/g ?

yeardf = pd.read_sas('NHANES2015_2016/DEMO_I.XPT')
Creatdf = pd.read_sas('NHANES2015_2016/ALB_CR_I.XPT')
serumCre = pd.read_sas('NHANES2015_2016/BIOPRO_I.XPT')
```

```python
serumCre.head()
```

```python
yeardf=yeardf.loc[:,['SEQN','RIAGENDR','RIDAGEYR','RIDRETH1']]
yeardf  = yeardf.dropna()
Creatdf=Creatdf.loc[:,['SEQN','URXUCR','URDACT']]
Creatdf  = Creatdf.dropna()
serumCre = serumCre.loc[:,['SEQN','LBXSCR']]
serumCre  = serumCre.dropna()
```

```python
df_temp = pd.merge(pd.merge(yeardf,Creatdf,how='inner',on='SEQN'),serumCre,how='inner',on='SEQN')
```

```python
df_temp.head()
```

```python
def eGFR(sCr,age,sex,race):
    if (sex == 2):
        
        if (race == 4) :
            gfr = 175 * m.pow((sCr),-1.154) * m.pow(age,-0.203) * 0.742 * 1.212
        else:
            gfr = 175 * m.pow((sCr),-1.154) * m.pow(age,-0.203) * 0.742
    else:
        
        if (race == 4) :
            gfr = 175 * m.pow((sCr),-1.154)* m.pow(age,-0.203) * 1.212
        else:
            gfr = 175 * m.pow((sCr),-1.154)* m.pow(age,-0.203) 
            
    return gfr        
```

```python
df_temp['GFR'] = df_temp.apply(lambda x : eGFR(x['LBXSCR'],x['RIDAGEYR'],x['RIAGENDR'],x['RIDRETH1']),axis=1)
```

```python
df_temp.head()
```

```python
def ckd(gfr,cratio):
    if((gfr<60) and (cratio>=30)):
        return 1
    else:
        return 2
    
```

```python
df_temp['KIQ022'] = df_temp.apply(lambda x : ckd(x['GFR'],x['URDACT']),axis=1)
```

```python
#Function to define the Class Hypertension
def hyp (sbp):
    #if (sbp < 140):
    if (sbp < 130):    
        return 0
    else:
        return 1
    
```
